```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('tweets_data_07_20.csv')
```

```python
opinions = list(np.unique(data['opinion'], return_counts=True))[0]
Counts = list(np.unique(data['opinion'], return_counts=True))[1]
```

```python
plt.xticks(rotation=90)
plt.title('General Public Sentiment on Covid19Trackers as of July20,2020')
plt.bar(opinions,Counts)
```
